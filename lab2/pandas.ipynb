### Сортировка данных

### Группировка данных
Pandas также позволяет группировать данные и выполнять агрегирующие операции.

```python
df_survived = df.groupby(['Pclass','Survived']).PassengerId.count().reset_index() 
df_survived = df_survived.merge(df_survived.groupby('Pclass').PassengerId.sum().reset_index(), on = 'Pclass')
df_survived.rename(columns={'PassengerId_x':'pass_count','PassengerId_y':'total_pass_count'},inplace=True)
df_survived['ratio'] = (df_survived['pass_count']*100/df_survived['total_pass_count']).round(2)
df_survived
```
